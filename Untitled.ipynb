```python
numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()
not_normal_features = []

for feature in numerical_features:
    stat, p = shapiro(df[feature])
    print(f'Shapiro-Wilk Test for {feature}: Statistics={stat}, p={p}')
    if p < 0.05:
        not_normal_features.append(feature)

# Apply Quantile Transformer to non-normal features
qt = QuantileTransformer(output_distribution='normal')
df[not_normal_features] = qt.fit_transform(df[not_normal_features])
print(f"Applied Quantile Transformer to: {not_normal_features}")
```

```python
df['Age_Category'] = pd.cut(df['Age (years)'], bins=[0, 2, 5, 10, 20], labels=['0-2', '2-5', '5-10', '10+'])
df['Capacity_Category'] = pd.cut(df['Capacity (BTUs)'], bins=[0, 10000, 20000, 30000], labels=['0-10k', '10k-20k', '20k-30k'])
```

```python
df['Age_Category'] = LabelEncoder().fit_transform(df['Age_Category'])
df['Capacity_Category'] = LabelEncoder().fit_transform(df['Capacity_Category'])
```

```python
label_encoders = {}
categorical_columns = ['Brand', 'Type', 'Features', 'Location']

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
```

```python
X = df.drop('Price ($)', axis=1)
y = df['Price ($)']
```

```python
iso = IsolationForest(contamination=0.1)
yhat = iso.fit_predict(X)
mask = yhat != -1
X, y = X[mask], y[mask]
```

```python
qt = QuantileTransformer(output_distribution='normal')
X_scaled = qt.fit_transform(X)
```

```python
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
X_poly = poly.fit_transform(X_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.2, random_state=42)
```

```python
param_grid_gbr = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5]
}
gbr = GradientBoostingRegressor()
grid_search_gbr = GridSearchCV(estimator=gbr, param_grid=param_grid_gbr, cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
grid_search_gbr.fit(X_train, y_train)
```

```python
# Get the best model
best_gbr = grid_search_gbr.best_estimator_

# Predict and evaluate
y_pred_gbr = best_gbr.predict(X_test)
print('Best Gradient Boosting Regressor RMSE:', np.sqrt(mean_squared_error(y_test, y_pred_gbr)))
print('Best Gradient Boosting Regressor R2:', r2_score(y_test, y_pred_gbr))
```

```python
# Train a Deep Learning model
model = Sequential()
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(1, activation='linear'))

model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

history = model.fit(X_train, y_train, validation_split=0.2, epochs=200, batch_size=16, verbose=1, callbacks=[early_stopping])
```

```python
# Predict and evaluate the Deep Learning model
y_pred_dl = model.predict(X_test)
print('Deep Learning Model RMSE:', np.sqrt(mean_squared_error(y_test, y_pred_dl)))
print('Deep Learning Model R2:', r2_score(y_test, y_pred_dl))
```
